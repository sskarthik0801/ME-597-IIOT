### Q1. Please summarize Lab10.

---

Lab 10 implemented a real-time machine sound monitoring system for a vacuum pump. Using an internal stethoscope/microphone sensor and MTConnect for audio streaming, a CNN model was deployed on a Raspberry Pi to classify the pump's operational state (OFF, Vacuuming, Air Leaking) from mel spectrogram features. Results were streamed via a local MTConnect adapter into MySQL and visualized in real time on a Grafana dashboard.
