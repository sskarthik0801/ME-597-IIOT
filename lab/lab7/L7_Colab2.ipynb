**<img src="https://github.com/purduelamm/purdue_me597_iiot/blob/main/lab/img/icon_SQL.png?raw=tru" width="40">SQL - SELECT statement for Practice #1**

```
SELECT
  timestamp AS "time",
  0+value as "Temperature"
FROM grafana_sample
WHERE
  $__timeFilter(timestamp) AND
  measurement = "temp" AND
  value != "UNAVAILABLE"
ORDER BY timestamp

```

---

**<img src="https://github.com/purduelamm/purdue_me597_iiot/blob/main/lab/img/icon_Tip.png?raw=tru" width="20">Tip - SELECT statement in Grafana**

1.	The reason why put ‘0+’ in front of ‘value’ column in the SELECT statement is to make the data in a numeric data type. We set up the ‘value’ column in the MySQL as ‘TINYTEXT’ that means basically return data is string data type.
2.	In addition, in the WHERE clause, ‘value != “UNAVAILABLE”’ makes your SELECT statement exclude “UNAVAILABE” data in the ‘value’ column.
3.	‘\$\__’ is indicating the macro function in Grafana to correlate the time range of Grafana interface with the database. ‘$__timeFilter(timestamp)’ means timestamps between the given time range. If you click ‘Show Help’ in the Query section, you will see the details of descriptions about the macro as below.

### Task 2.3

  1. Is it possible to build the IIoT monitoring system without MTConnect? Discuss the pros and cons of using MTConnect in the system.



---

Yes, it is possible to build an IIoT monitoring system without MTConnect by directly collecting sensor data using custom programs and sending the data to a database.

Pros of using MTConnect

- Provides a standardized protocol for machine data communication.

- Makes integration of different machines and sensors easier.

- Uses a structured XML format that is consistent and machine-readable.

- Improves interoperability between devices from different vendors.

Cons of using MTConnect

- Adds extra system components such as adapters and agents.

- Increases system complexity compared to direct data collection.

- XML processing may introduce additional processing overhead.
