## 1.3 Load Machine Learning Model Using TensorFlow on Raspberry Pi

Let’s load the model on Raspberry Pi you finally developed in Lab8. Please follow the steps below.

1.	Make a working directory in which your model and Python file will be.

  a.	The directory example: ‘/home/pi/prelab9/models’

<img src="https://github.com/purduelamm/purdue_me597_iiot/blob/main/lab/img_S26_update/Prelab9_fig11.png?raw=tru" width="90%">

2.	Copy and paste the best model directory you created in Lab 8 to Raspberry Pi directory
(‘/home/pi/prelab9/models’)

  a.	Check the model folder is moved to the Raspberry Pi

  b.	In this example, the model folder name is ‘*20260115_222313_lab8_anomaly_x-axis_raw’*.

<img src="https://github.com/purduelamm/purdue_me597_iiot/blob/main/lab/img_S26_update/Prelab9_fig12.png?raw=tru" width="90%">

3.	Create a Python file as ‘prelab9_load_model.py’ on the working directory. In this example, the working directory is ‘/home/pi/prelab9/models/’.



4.	Code the Python script below in the created Python file (‘prelab9_load_model.py’) and then save it.

---

**<img src="https://github.com/purduelamm/purdue_me597_iiot/blob/main/lab/img/icon_Python.png?raw=tru" width="20">Python - Python3 (prelab9_load_model.py)**

```
import tensorflow as tf

MODEL_DIR = "/home/pi/prelab9/models/20260115_232213_lab8_anomaly_x-axis_raw"

loaded = tf.saved_model.load(MODEL_DIR)
infer = loaded.signatures["serving_default"]

_ = infer(tf.zeros((1, 1000), dtype=tf.float32))

print("Model load success.")

```
