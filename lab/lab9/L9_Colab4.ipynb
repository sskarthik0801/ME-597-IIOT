## Lab9 Summary and Deliverables

### Q1. Please summarize Lab9.

---

in this lab, an IIoT-based smart monitoring system was developed for an axial flow fan (AFF). The system integrates sensor data acquisition using the ADXL345 accelerometer, machine learning-based anomaly detection, and real-time data streaming using MTConnect. The execution state of the system was determined using an RMS threshold, and the condition (normal or abnormal) was predicted using a trained autoencoder model. The collected data was then stored in a MySQL database and visualized through a Grafana dashboard, enabling real-time monitoring and analysis.
